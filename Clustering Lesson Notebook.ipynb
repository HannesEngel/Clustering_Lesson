## Welcome to the Clustering task!

Previously, we discussed the difference between supervised and unsupervised machine learning techniques. Recall that for supervised learning techniques, the machine predicts an output for an input by learning from pairs of labelled inputs and outputs; that is, the program learns from examples that have the right answers assigned to them beforehand. For unsupervised learning, however, the program does not learn from explicitly labelled data. Instead, unsupervised learning draws inferences from input data without labelled output data, i.e. it aims to discover patterns in the data on its own, without human supervision.

### Introduction to clustering:

A clustering analysis can be used to ascertain, on the basis of your input variables, whether the observations in the data fall into relatively distinct groups with respect to the input variables/features. The goal of such an exercise would be to split the data points into its distinct groups; that is, we try to classify each data point into a specific group. The ideal outcome of such an exercise would be to have observations within each group that are similar, while observations in different groups are quite different from each other.

For example, suppose we were tasked with analysing Canadian census data. This data includes information about the demographic makeup of various provinces/cities in Canada. # 2. K-Means Clustering

K-Means Clustering is a very popular clustering algorithm used in practice today. It is a simple and elegant approach to partitioning data into K distinct groups. K-means works as follows:

You select K, a predetermined number of clusters that you want to form. Then K points (centroids) are selected at random locations in feature space. For each point in your training data set:
1. You find the centroid that the point is closest to. **
2. And assign that point to that cluster.
3. Then, for each cluster centroid, you move that point such that it is in the centre of all the points that are were assigned to that cluster in step 2.
4. Repeat steps 2 and 3 until youâ€™ve either reached convergence and points no longer change cluster membership, or until some specified number of iterations has been reached.

# 3. Validating clusters

Once our clustering algorithm has been run, we will want to find out how our algorithm has partitioned the data. Recall that our aim is to have a small within-cluster variation (values in the same cluster differ by a small amount) but a high without-cluster variation (values in different clusters differ by larger amounts). 

One way of evaluating this is to use the Silhouette Coefficient, which gives us an indication of how close each observation is to its own (final) cluster. The Silhouette Coefficient varies between -1 and 1, with values close to 1 indicating the selection of distinct clusters and values close to -1 indicating overlapping clusters. Scikit-Learnâ€™s documentation on the Silhouette Coefficient can be found here: 

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html#sklearn.metrics.silhouette_score

# 4. Choosing the value of K

Determining the number of clusters (K) to use in your algorithm is not always a straightforward exercise. It will likely depend on two factors:

1.	The specifics and domain knowledge of the problem you are trying to solve. When you are working on a familiar problem, it might be that you have a relatively good idea of the number of clusters that may arise from the data. In such cases, your initial 'gut-feel' for the value of K might serve as a good starting point. However, for high-dimensional data this is often not the case.
2.	The K-Means algorithmâ€™s â€˜performanceâ€™ (e.g. Silhouette Coefficient) on a set of different values for K. Often, the value of K is varied over a wide range of possible values. The K-Means algorithm is then fit to the data for each of these values of K, with the Silhouette Coefficient being measured each run. The value of K that gives the highest Silhouette Coefficient is selected as the most effective and final value for K. This approach will be discussed in a more practical setting in section 5 below.
